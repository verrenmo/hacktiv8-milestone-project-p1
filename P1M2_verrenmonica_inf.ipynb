# 2. Import Libraries

```python
# Import library
import pickle
import pandas as pd
```

# 3. Model Loading

```python
# Open the best model
with open('bestModel.pkl', 'rb') as modelFile:
  bestModel = pickle.load(modelFile)
```

# 4. Inference Data

```python
# Input inference data
data={ 
       'person_age': [31, 27, 22, 86, 50],
       'person_gender': ['male','female','female','male','male'],
       'person_education': ['Master','High School','Associate','Doctorate','Bachelor'],
       'person_income': [23744, 47981, 14532, 71947, 36689],
       'person_emp_exp' :[10, 6, 1, 55, 23],
       'person_home_ownership': ['RENT','MORTGAGE','OTHER','OWN','OWN'],
       'loan_amnt': [4200, 20000, 1500, 9000, 10000],
       'loan_intent': ['PERSONAL','MEDICAL','EDUCATION','MEDICAL','HOME IMPROVEMENT'],
       'loan_int_rate':[11.14, 10.7, 10, 15, 12],
       'loan_percent_income':[0.17, 0.41, 0.10, 0.125, 0.27],
       'cb_person_cred_hist_length':[10, 5 , 1, 30, 11],
       'credit_score':[432, 305, 200, 678, 504],
       'previous_loan_defaults_on_file':['No','Yes','No','No','Yes']
       }

dataInf = pd.DataFrame(data)

dataInf
```

# 5. Preprocessing

Pada bagian ini dilakukan perubahan nama kolom.

```python
# Rename columns name
dataInf.rename(columns={
             'person_age':'age',
             'person_gender':'gender',
             'person_education':'education',
             'person_income':'income',
             'person_emp_exp':'empExp',
             'person_home_ownership':'homeOwnership',
             'loan_amnt':'loanAmount',
             'loan_intent':'loanIntent',
             'loan_int_rate':'intRate',
             'loan_percent_income':'loanPctIncome',
             'cb_person_cred_hist_length':'creditHistLen',
             'credit_score':'creditScore',
             'previous_loan_defaults_on_file':'prevDefault'
             }, inplace=True)
dataInf.head(5)
```

# 6. Prediction Result

Pada bagian ini dilakukan prediksi data inference.

```python
# Prediksi dengan inference data
pred = bestModel.predict(dataInf)

predReslt = []
for i in pred:
    if i == 0:
        predReslt.append("Rejected")
    elif i == 1:
        predReslt.append("Approved")

dataInf['predictResult'] = predReslt
dataInf
```

Dari hasil prediksi data inference dengan model yang telah dibuat, maka pemohon 1, 3, dan 4 pengajuannya diterima (approved) dan pemohon 2 dan 5 pengajuannya ditolak (rejected).
